### Handling Outliers

```python
df_sub02.describe()
```

```python
def detect_outliers(df, x):
    Q1 = df[x].describe()['25%']
    Q3 = df[x].describe()['75%']
    IQR = Q3-Q1
    return df[(df[x] < Q1-1.5*IQR) | (df[x] > Q3+1.5*IQR)]
```

```python
#dati2
detect_outliers(df_sub02, 'dati2')
```

```python
dati2_outliers = list(detect_outliers(df_sub02, 'dati2').index)
```

```python
len(list(detect_outliers(df_sub02, 'dati2').index))
```

```python
for i in dati2_outliers:
    df_sub02.drop([i], inplace = True )
```

```python
#usia
len(list(detect_outliers(df_sub02, 'usia').index))
```

## Resampling

```python
df_encoding['label'].value_counts()
```

```python
#visualisasi label dengan pie chart
df_encoding['label'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True)
```

**Ket:** Data very very imbalance

```python
X = df_encoding.drop(['label'], axis=1)
```

```python
X = X.drop(['id', 'id_peserta', 'diag', 'levelid'], axis=1)
```

```python
X
```

```python
y = df_encoding['label']
```

```python
y
```

**Oversampling data using SMOTE**

```python
#Metode SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=2)
X_sm, y_sm = smote.fit_resample(X, y)
```

```python
#Vizualise label using pie chart
y_sm.value_counts().plot.pie(autopct='%1.1f%%', shadow=True)
```

**Data sudah balance**

```python
from sklearn.model_selection import train_test_split
```

## Model Building

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=1, stratify=y_sm)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_predict)
print("Precision: %.2f%%" % (precision * 100.0))
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_predict)
print("Recall: %.2f%%" % (recall * 100.0))
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_predict)
print("F1: %.2f%%" % (f1 * 100.0))
```

**Note:** Model final (using all of dataset)

```python
model_final = XGBClassifier()
model_final.fit(X_sm, y_sm)
```
